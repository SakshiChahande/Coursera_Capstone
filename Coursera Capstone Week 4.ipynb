## Data Selection

I am comparing India's two largest cities, Mumbai and Delhi.

I have obtained the data-set from the following link:

https://data.gov.in/resources/all-india-pincode-directory-contact-details-along-latitude-and-longitude/api

This dataset contains the postcodes of the post offices of each locality in India. I have extracted  two dataframes from this dataset and applied the geopy api to find and integrate the latitude and longitude for each area. This has helped me create the map using Folium api.

```python
url = r'C:\Users\saksh\Documents\Coursera\DS_Capstone\all_india_PO_list_without_APS_offices_ver2_lat_long.csv'
df=pd.read_csv(url)
df.head()
```

```python
sizeoftable=df.shape[0]
sizeoftable
```

```python
dfDelhi=df[df.divisionname.str.contains("Delhi")]
dfDelhi=dfDelhi.reset_index(drop=True)
dfDelhi
```

```python
delhiArea=dfDelhi.shape[0]
```

```python
i=0
while(i<delhiArea):
    address=dfDelhi['officename'][i] + ',' +dfDelhi['statename'][i] + ',India' 
    geolocator = Nominatim(user_agent="my-application")
    location = geolocator.geocode(address)
    if(not location == None):
        latitude = location.latitude
        longitude =location.longitude
        dfDelhi['latitude'][i]=latitude
        dfDelhi['longitude'][i]=longitude
    i=i+1
    address=None
    location=None
    latitude=None
    longitude=None
```

```python
dfDelhi.dtypes
```

```python
dfMumbai=df[df.divisionname.str.contains("Mumbai")]
dfMumbai=dfMumbai.reset_index(drop=True)
dfMumbai
```

```python
mumbaiArea=dfMumbai.shape[0]
mumbaiArea
```

```python
j=0
while(j<mumbaiArea):
    address=dfMumbai['officename'][j] + ',' +dfMumbai['statename'][j] + ',India' 
    geolocator = Nominatim(user_agent="my-application")
    location = geolocator.geocode(address)
    if( not location == None):
        latitude = location.latitude
        longitude =location.longitude
        dfMumbai['latitude'][j]=latitude
        dfMumbai['longitude'][j]=longitude
        
    j=j+1
    address=None
    location=None
    latitude=None
    longitude=None
```

```python
dfMumbai1=dfMumbai['latitude'].dropna()
```

```python
Mumbai=pd.merge(dfMumbai,dfMumbai1, how='right', on = 'latitude')
Mumbai.head(15)
```

```python
dfDelhi1=dfDelhi['latitude'].dropna()
```

```python
Delhi=pd.merge(dfDelhi,dfDelhi1, how='right', on = 'latitude')
Delhi.head(15)
```

```python
address='Delhi, India'
geolocator = Nominatim(user_agent="my-application")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Delhi are {}, {}.'.format(latitude, longitude))
```

```python

# create map of Delhi using latitude and longitude values
map_Delhi = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, label in zip(Delhi['latitude'], Delhi['longitude'], Delhi['officename']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        popup=label,
        color='black',
        fill=True,
        fill_color='black',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Delhi)  
    
    
map_Delhi
```

```python
address='Mumbai, India'
geolocator = Nominatim(user_agent="my-application")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Mumbai are {}, {}.'.format(latitude, longitude))
```

```python

# create map of Delhi using latitude and longitude values
map_Mumbai = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, label in zip(Mumbai['latitude'], Mumbai['longitude'], Mumbai['officename']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        popup=label,
        color='purple',
        fill=True,
        fill_color='purple',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Mumbai)  
    
    
map_Mumbai
```
